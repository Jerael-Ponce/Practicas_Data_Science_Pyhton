```python
'''
10 - Un programa debe ser escrito para leer dos números y luego preguntar a
la persona usuaria qué operación desea realizar. El resultado de la operación
debe incluir información sobre el número, si es par o impar, positivo o
 negativo, e entero o decimal.
'''
def informacionNumero(reulstado):
  print(f'El resultado es {reulstado}')
  if reulstado%2 == 0:
    print(f'Es par')
  else:
    print(f'Es impar')
  if reulstado > 0:
    print(f'Es positivo')
  else:
    print(f'Es negativo')
  if reulstado == int(reulstado):
    print(f'Es entero')
  else:
    print(f'Es decimal')


primerNumero = float(input('Ingrese un numero: '))
segundoNumero = float(input('Ingrese otro numero: '))
operacion = input(f'Escoja entre alguna de las siguientes operaciones\n A) suma \n B) resta \n C) multiplicacion \n D) division \n')
if operacion.upper() == 'A':
  reulstado = primerNumero+ segundoNumero
  informacionNumero(reulstado)
elif operacion.upper() == 'B':
  reulstado = primerNumero- segundoNumero
  informacionNumero(reulstado)
elif operacion.upper() == 'C':
  reulstado = primerNumero* segundoNumero
  informacionNumero(reulstado)
elif operacion.upper() == 'D':
  reulstado = primerNumero/ segundoNumero
  informacionNumero(reulstado)
else:
  print('Ingrese una operacion valida')
```

```python
'''
11 - Escribe un programa que pida a la persona usuaria tres números que
representan los lados de un triángulo. El programa debe informar si los
valores pueden utilizarse para formar un triángulo y, en caso afirmativo,
si es equilátero, isósceles o escaleno. Ten en cuenta algunas sugerencias:

Tres lados forman un triángulo cuando la suma de cualesquiera dos lados es mayor que el tercero;
Triángulo Equilátero: tres lados iguales;
Triángulo Isósceles: dos lados iguales;
Triángulo Escaleno: tres lados diferentes.
'''
lado_1 = float(input('Ingrese el primer lado del triangulo: '))
lado_2 = float(input('Ingrese el segundo lado del triangulo: '))
lado_3 = float(input('Ingrese el tercer lado del triangulo: '))

if (lado_1 + lado_2 >lado_3) or (lado_1 + lado_3 > lado_2) or (lado_2 + lado_3 > lado_1):
  print('Los valores ingresados pueden formar un triangulo')
  if (lado_1 == lado_2 == lado_3):
    print('Puede ser un triangulo equilatero')
  elif (lado_1 == lado_2) or (lado_1 == lado_3) or (lado_2 == lado_3):
    print('Puede ser un triangulo isoceles')
  else:
    print('Puede ser un triangulo escaleno')
else:
  print('Los valores ingresados no pueden formar un triangulo')
```

```python
'''
12 - Un establecimiento está vendiendo combustibles con descuentos variables.
Para el etanol, si la cantidad comprada es de hasta 15 litros, el descuento
será del 2% por litro. En caso contrario, será del 4% por litro. Para el diésel,
 si la cantidad comprada es de hasta 15 litros, el descuento será del 3% por
litro. En caso contrario, será del 5% por litro. El precio por litro de diésel
es de R$ 2,00 y el precio por litro de etanol es de R$ 1,70. Escribe un
programa que lea la cantidad de litros vendidos y el tipo de combustible
(E para etanol y D para diésel) y calcule el valor a pagar por el cliente.
Ten en cuenta algunas sugerencias:

El valor del descuento será el producto del precio por litro, la cantidad de litros y el valor del descuento.
El valor a pagar por un cliente será el resultado de la multiplicación del precio por litro por la cantidad de litros menos el valor del descuento resultante del cálculo.

'''
precioEtanol = 1.7
precioDiese= 2.00
def calculoDescuento(factorDescuento,precioPorLitro,cantidadCombistustible):
  descuento=factorDescuento*precioPorLitro*cantidadCombistustible
  pago=precioPorLitro*cantidadCombistustible-descuento
  print(f'El cliente paga ${pago} con un descuento de ${descuento} incluido.')

combustibleConsumido = input('Conbustible a vender (E para etanol y D para diesel)')
litrosConsumidos = float(input('Litros a vender: '))
if combustibleConsumido.upper()=='E':
  if litrosConsumidos <= 15:
    calculoDescuento(0.02,precioEtanol,litrosConsumidos)
  else:
    calculoDescuento(0.04,precioEtanol,litrosConsumidos)
elif combustibleConsumido.upper()=='D':
  if litrosConsumidos <= 15:
    calculoDescuento(0.03,precioDiese,litrosConsumidos)
  else:
    calculoDescuento(0.05,precioDiese,litrosConsumidos)
else:
  print('Ingrese un tipo de combustible valido')
```

```python
'''
13 - En una empresa de venta de bienes raíces, debes crear un código que
analice los datos de ventas anuales para ayudar a la dirección en la toma de
decisiones. El código debe recopilar los datos de cantidad de ventas durante
los años 2022 y 2023 y calcular la variación porcentual. A partir del valor de
la variación, se deben proporcionar las siguientes sugerencias:

Para una variación superior al 20%: bonificación para el equipo de ventas.
Para una variación entre el 2% y el 20%: pequeña bonificación para el equipo de ventas.
Para una variación entre el 2% y el -10%: planificación de políticas de incentivo a las ventas.
Para bonificaciones inferiores al -10%: recorte de gastos.
'''
ventas2022 = float(input('Ingrese la canctidad de ventas del 2022: '))
ventas2023 = float(input('Ingrese la canctidad de ventas del 2023: '))
variacion = (ventas2023-ventas2022)/ventas2022
if variacion > 0.2:
  print('Bonificacion para el equipo de ventas')
elif variacion >= 0.02 and variacion <= 0.2:
  print('Pequeña bonificacion para el equipo de ventas')
elif variacion >= -0.1 and variacion < 0.02:
  print('Planificacion de políticas de incentivo a las ventas')
else:
  print('Recorte de gastos')
```
