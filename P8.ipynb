1 - Haz un programa que solicite a la persona usuaria ingresar dos números decimales y calcular la división entre estos números. El código debe incluir un manejo de error, indicando el tipo de error que se generó si la división no es posible.

Prueba el programa con el segundo valor numérico de la entrada igual a 0. También prueba usando caracteres textuales en la entrada para verificar los tipos de errores que ocurren.

````python
print('Este programa realiza la division entre dos numeros decimales')
try:
  numero1 = float(input('Ingresa el primer numero: '))
  numero2 = float(input('Ingresa el segundo numero: '))
except ValueError:
  print('Ingrese solo numeros, por favor')
except Exception as e:
  print(type(e),e)
else:
  try:
    division = numero1/numero2
  except ZeroDivisionError:
    print('No es posible una division por cero')
  except Exception as e:
    print(type(e),e)
  else:
    print(f'El resultado es: ${division}')
finally:
  print('Proceso finalizado')
````

4 - Crea una función que reciba dos listas como parámetros y agrupe los elementos uno a uno de las listas, formando una lista de tuplas de 3 elementos. El primer y segundo elemento de la tupla son los valores en la posición i de las listas y el tercer elemento es la suma de los valores en la posición i de las listas.

La función debe incluir un manejo de error indicando el tipo de error generado y devolver como resultado la lista de tuplas. Si las listas enviadas como parámetro tienen tamaños diferentes, la función debe devolver un IndexError con la frase: 'La cantidad de elementos en cada lista es diferente.'.

Datos para probar la función:

Valores sin error:

```
lista1 = [4, 6, 7, 9, 10]
lista2 = [-4, 6, 8, 7, 9]
```

Listas con tamaños diferentes:

```
lista1 = [4, 6, 7, 9, 10, 4]
lista2 = [-4, 6, 8, 7, 9]
```

Listas con valores incoherentes:


````python
#Listas con tamaños diferentes:
lista1 = [4, 6, 7, 9, 10, 4]
lista2 = [-4, 6, 8, 7, 9]
````

````python
#Listas con valores incoherentes:
lista1 = [4, 6, 7, 9, 'A']
lista2 = [-4, 'E', 8, 7, 9]
````

````python
nueva_lista=list2tupla(lista1,lista2)
print(nueva_lista)
````

````python
def list2tupla(lista1: list=[0],lista2: list=[0])->list:
  '''
  El primer y segundo elemento de la tupla son los valores en la posición
  i de las listas y el tercer elemento es la suma de los valores en la
  posición i de las listas.

  lista1: list, default [0]
    Primer lista a operar
  lista2: list, default [0]
    Segunda lista a operar

  return new_list: list
    Lista con forma
    new_list=[([0]1,[0]2,[0]1+[0]2),...,([n-1]1,[n-1]2,[n-1]1+[n-1]2)]
  '''
  try:
    if len(lista1) != len(lista2):
      raise IndexError('La cantidad de elementos en cada lista es diferente.')
    new_list=[(i,j,i+j) for i,j in zip(lista1,lista2)]
  except TypeError:
    print('No se pueden convertir string a float')
    return None
  except IndexError as e:
    print(e)
  else:
    return new_list
````

````python
def calculo_calificacion(lista_calificaciones: list=0, alternativas: list=[0],respuestas: list=[0])->list:
  '''
  Funcion que calcula la calificacion de los alumnos en una preuba de opcion multiple

  lista_calificaciones: lista default = [0]
    Lista de listas donde cada sublista tiene todas las calificaiones correspondientes
    a la prueba de un solo alumno
    [calificaciones alumno 1, calificaciones alumno 2, ..., calificaiones alumno n]

  alternativas: lista default = [0]
    Lista con las posbles respues que estan presentes en la prubea

  respuesta: lista defaut=[0]
    Lista con las respuestas de todas las preguntas de la prueba

  resultado_calificnaciones:
    Lista con las calificaciones resultantes para cada alumno
  '''

  try:
    resultado_calificnaciones=[]
    for i,respuestasEstudiantes in enumerate(lista_calificaciones):
      resultado_calificnaciones.append(0)
      for j,respuesta in enumerate(respuestasEstudiantes):
        if respuesta in alternativas:
          if respuesta == respuestas[j]:
            resultado_calificnaciones[i] +=1
        else:
          raise ValueError(f'La alternativa {respuesta} no es una opción de alternativa válida')
  except ValueError as e:
    print (e)
  else:
    return resultado_calificnaciones
  finally:
    print('Proceso finalizado')

````

````python
def verificacionPalabras(listaDePalabras: list=[])->str:
  caraccteresEspeciaes=[',', '.', '!' , '?']
  for palabra in listaDePalabras:
    try:
      for caracter in caraccteresEspeciaes:
        if caracter in palabra:
          raise ValueError(f'El texto presenta puntuaciones en la palabra {palabra}')
    except ValueError as e:
      print(type(e),e)
````
