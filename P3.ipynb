```python
'''
1 - Escribe un programa que solicite dos números enteros e imprima todos los números enteros entre ellos.
'''
print('Vamos a imprimir todos los numeros que esten entre el rango que indiques')
incio = float(input('Ingresa el numero de inicio: '))
final = float(input('Ingresa el numero final: '))
if final < incio:
    nuevoInicio = final
    final = incio
    incio = nuevoInicio
    exit()
for numero in range(int(incio), int(final) + 1):
    print(numero)
```

```python
'''
2 - Escribe un programa para calcular cuántos días tomará que la colonia de una
bacteria A supere o iguale a la colonia de una bacteria B, basado en tasas de
crecimiento del 3% y 1.5%, respectivamente. Supón que la colonia A comienza con
4 elementos y B con 10.
'''
crecimientoA = 0.03
crecimientoB= 0.015
dias = 1
bacteriasA = 4
bacteriasB = 10
while bacteriasA <= bacteriasB:
  bacteriasA += bacteriasA * crecimientoA
  bacteriasB += bacteriasB * crecimientoB
  dias += 1
print(f'Tomara {dias} dias para que la bacteria A iguale o supere a la bacteria B')
```

```python
'''
3 - Para procesar una cantidad de 15 datos de evaluaciones de usuarios de un
servicio de la empresa, necesitamos verificar si las calificaciones son válidas.
Por lo tanto, escribe un programa que recibirá calificaciones del 0 al 5 y
verificará si son valores válidos. Si se ingresa una calificación superior a 5
o inferior a 0, se repetirá hasta que el usuario ingrese un valor válido.
'''
numeroEvaluacion = 1
while numeroEvaluacion <= 15:
  evaluacion = float(input(f'Ingresa la evaluacion {numeroEvaluacion} con un numero entre el 0 y 5: '))
  if evaluacion<0 or evaluacion>5 :
    print(f'Ingrese una calificacion valida para a evaluacion {numeroEvaluacion}')
  else:
    print(f'La evaluacion {numeroEvaluacion}  con un puntaje de {evaluacion} es valida')
    numeroEvaluacion += 1
```

```python
'''
4 - Desarrolla un programa que lea un conjunto indefinido de temperaturas en
grados Celsius y calcule su promedio. La lectura debe detenerse al ingresar el
valor -273°C.
'''
print(f'Este programa calculara el promedio de las temperaturas que agregues. \n Para detener el proceso -273')

contador = 1
promedio = 0
cuenta = 0
while True:
  temperatura= (input(f'Ingresa la temperatura {contador}: '))
  if float(temperatura):
    temperatura = float(temperatura)
    if temperatura == -273:
      break
    else:
      cuenta += temperatura
      promedio = cuenta / contador
      contador += 1
      print(f'El promedio acumulado es {promedio}')
  else:
    continue

print('Acabamos por el momento')
```

```python
'''
5 - Escribe un programa que calcule el factorial de un número entero
proporcionado por el usuario. Recuerda que el factorial de un número entero es
el producto de ese número por todos sus antecesores hasta llegar al número 1.
 Por ejemplo, el factorial de 5 es 5 x 4 x 3 x 2 x 1 = 120.
'''
cuenta = 1
numeroFactorial = int(input('Ingrese un numero entere que desee saber su factorial: '))
for numero in range(1,numeroFactorial+1):
  cuenta *= numero
print(f'El resultadi factorial de {numeroFactorial} es {cuenta}')
```

```python
'''
7 - Los números primos tienen diversas aplicaciones en Ciencia de Datos, como
en criptografía y seguridad. Un número primo es aquel que es divisible solo por
sí mismo y por 1. Por lo tanto, crea un programa que solicite un número entero
y determine si es un número primo o no.
'''
numero = int(input('Ingresa un numero entero para saber si es primo o no: '))
print(f'El numero {numero}...')
for multiplo in range (2,numero+1):
  if multiplo != numero and numero % multiplo != 0:
    print(f'No es multiplo de {multiplo}.')
    continue
  elif multiplo != numero and numero % multiplo ==0:
    print(f'El numero {numero} no es primo')
    break
  else:
    print(f'El numero {numero} es primo')
```

```python
'''
8 - Vamos a comprender la distribución de edades de los pensionistas de una
empresa de seguros. Escribe un programa que lea las edades de una cantidad no
informada de clientes y muestre la distribución en los intervalos [0-25],
[26-50], [51-75] y [76-100]. La entrada de datos se detendrá al ingresar un
número negativo.
'''
primerCuartil= 0
segundoCuartil= 0
tercerCuartil= 0
cuartoCuartil= 0
while True:
  edad=int(input('Ingresa tu edad:'))
  if edad < 0:
    break
  else:
    if edad >= 0 and edad <= 25:
      primerCuartil += 1
    elif edad >= 26 and edad <= 50:
      segundoCuartil += 1
    elif edad >= 51 and edad <= 75:
      tercerCuartil += 1
    elif edad >= 76 and edad <= 100:
      cuartoCuartil += 1
    print(f'[0-25]: {primerCuartil}\t[26-50]: {segundoCuartil}\t[51-75]: {tercerCuartil}\t[76-100]: {cuartoCuartil}')
```
